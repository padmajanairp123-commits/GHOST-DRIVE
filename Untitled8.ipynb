## Simulate Virtual LiDAR Data

### Subtask:
Generate synthetic LiDAR scan data within a simulated environment. This step will involve creating a virtual scene with obstacles and a moving 'robot' equipped with a LiDAR sensor, simulating how the sensor would perceive the environment.


**Reasoning**:
The first step is to import the necessary libraries for numerical operations and plotting, and then define the virtual environment, obstacles, and the robot's initial state as outlined in the instructions.



```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the virtual environment
environment_size = 50 # meters (e.g., a 50x50 meter square environment)

# 2. Define obstacles (simple circles for now: [x, y, radius])
obstacles = [
    [10, 70, 3],  # Obstacle 1: x=10, y=10, radius=3
    [25, 20, 5],  # Obstacle 2: x=25, y=20, radius=5
    [40, 35, 4],  # Obstacle 3: x=40, y=35, radius=4
    [15, 15, 2]   # Obstacle 4: x=15, y=40, radius=2
]

# 3. Define the robot's initial position, orientation, and a simple movement trajectory
robot_pos = np.array([5.0, 5.0]) # Initial x, y position
robot_orientation = np.pi / 4 # Initial orientation (radians, 45 degrees)

# Store initial setup details
print("Virtual Environment Size:", environment_size, "m")
print("Obstacles (x, y, radius):")
for obs in obstacles:
    print(f"  - x={obs[0]}, y={obs[1]}, r={obs[2]}")
print("Robot Initial Position:", robot_pos)
print("Robot Initial Orientation:", np.degrees(robot_orientation), "degrees")
```

**Reasoning**:
Now that the environment, obstacles, and robot's initial state are defined, the next logical step is to implement the virtual LiDAR sensor model, including its parameters and a function to simulate the rays and detect intersections with obstacles. This will satisfy instructions 5 and 6 of the subtask.



```python
max_range = 30 # Maximum LiDAR detection range
num_beams = 360 # Number of LiDAR beams (360 degrees)
angular_resolution = 2 * np.pi / num_beams # Radians per beam

# Function to simulate a single LiDAR ray and detect intersection with obstacles
def simulate_lidar_ray(robot_pos, robot_orientation, angle_offset, obstacles, max_range):
    # Calculate the global angle of the current ray
    ray_angle = robot_orientation + angle_offset

    # Define the ray's start and end points for maximum range
    ray_start = robot_pos
    ray_end_x = robot_pos[0] + max_range * np.cos(ray_angle)
    ray_end_y = robot_pos[1] + max_range * np.sin(ray_angle)

    min_distance = max_range

    for obs in obstacles:
        obs_x, obs_y, obs_r = obs
        # Vector from ray start to obstacle center
        oc = np.array([obs_x - ray_start[0], obs_y - ray_start[1]])
        # Ray direction vector
        ray_dir = np.array([np.cos(ray_angle), np.sin(ray_angle)])

        # Project oc onto ray_dir to find t (distance along ray_dir to closest point to obstacle center)
        t = np.dot(oc, ray_dir)

        if t < 0: # Obstacle is behind the ray origin, check if origin is inside
            dist_to_center = np.linalg.norm(oc)
            if dist_to_center < obs_r:
                # Robot is inside the obstacle, this is a hit at 0 distance
                min_distance = 0.001 # Small positive distance to indicate hit
                break
            continue

        # Closest point on ray to obstacle center
        closest_point_on_ray = ray_start + t * ray_dir

        # Distance from obstacle center to closest point on ray
        dist_to_ray = np.linalg.norm(np.array([obs_x, obs_y]) - closest_point_on_ray)

        if dist_to_ray <= obs_r: # Intersection occurs
            # Calculate actual intersection distance from robot position
            # Using Pythagorean theorem: d_intersect = t - sqrt(r^2 - dist_to_ray^2)
            dist_from_closest_point = np.sqrt(obs_r**2 - dist_to_ray**2)
            actual_distance = t - dist_from_closest_point

            if actual_distance < min_distance and actual_distance >= 0:
                min_distance = actual_distance

    return min_distance

print(f"LiDAR parameters: Max Range = {max_range}m, Number of Beams = {num_beams}, Angular Resolution = {np.degrees(angular_resolution):.2f} degrees")

# Test the function with a single ray (e.g., straight ahead)
test_angle_offset = 0 # Directly in front of the robot
test_distance = simulate_lidar_ray(robot_pos, robot_orientation, test_angle_offset, obstacles, max_range)
print(f"Test LiDAR ray (angle offset {np.degrees(test_angle_offset)} deg) detected distance: {test_distance:.2f} m")
```

**Reasoning**:
With the `simulate_lidar_ray` function established, the next step is to create a function that performs a full LiDAR scan by calling this function for each beam. This function will collect all distance and angle measurements, addressing instruction #7 of the subtask.



```python
def generate_lidar_scan(robot_pos, robot_orientation, obstacles, max_range, num_beams, angular_resolution):
    scan_data = [] # List to store (distance, angle) pairs

    # Iterate through each beam angle
    for i in range(num_beams):
        # Calculate the angle offset for the current beam relative to the robot's orientation
        angle_offset = i * angular_resolution - np.pi # -np.pi to center the scan around 0 or robot's forward direction

        # Simulate the ray and get the distance
        distance = simulate_lidar_ray(robot_pos, robot_orientation, angle_offset, obstacles, max_range)

        # Store the distance and the local angle (relative to robot's forward)
        scan_data.append((distance, angle_offset))

    return np.array(scan_data)

# Generate a full LiDAR scan
lidar_scan = generate_lidar_scan(robot_pos, robot_orientation, obstacles, max_range, num_beams, angular_resolution)

print(f"Generated LiDAR scan with {len(lidar_scan)} beams.")
print("First 5 scan points (distance, angle_offset in degrees):")
for i in range(min(5, len(lidar_scan))):
    print(f"  Distance: {lidar_scan[i, 0]:.2f}m, Angle: {np.degrees(lidar_scan[i, 1]):.2f} deg")
```

**Reasoning**:
With the LiDAR scan data generated, the next step is to visualize the virtual environment, the robot's position, the obstacles, and the emitted LiDAR rays hitting the obstacles. This will satisfy instruction #8 of the subtask and allow for verification of the simulation setup.



```python
plt.figure(figsize=(10, 10))
plt.xlim(0, environment_size)
plt.ylim(0, environment_size)
plt.gca().set_aspect('equal', adjustable='box')

# Plot environment boundaries
plt.plot([0, environment_size], [0, 0], 'k-')
plt.plot([0, environment_size], [environment_size, environment_size], 'k-')
plt.plot([0, 0], [0, environment_size], 'k-')
plt.plot([environment_size, environment_size], [0, environment_size], 'k-')

# Plot obstacles
for obs in obstacles:
    circle = plt.Circle((obs[0], obs[1]), obs[2], color='red', alpha=0.6)
    plt.gca().add_patch(circle)
    plt.text(obs[0], obs[1], f'O{obstacles.index(obs)+1}', color='black', ha='center', va='center')

# Plot robot position and orientation
plt.plot(robot_pos[0], robot_pos[1], 'go', markersize=10, label='Robot Position')
robot_arrow_length = 2 # Length of the arrow indicating orientation
plt.arrow(robot_pos[0], robot_pos[1],
          robot_arrow_length * np.cos(robot_orientation),
          robot_arrow_length * np.sin(robot_orientation),
          head_width=0.8, head_length=0.8, fc='green', ec='green')

# Plot LiDAR scan points
# Convert polar coordinates (distance, angle_offset) to Cartesian (x, y)
scan_points_x = robot_pos[0] + lidar_scan[:, 0] * np.cos(robot_orientation + lidar_scan[:, 1])
scan_points_y = robot_pos[1] + lidar_scan[:, 0] * np.sin(robot_orientation + lidar_scan[:, 1])

# Plot the scan points
plt.plot(scan_points_x, scan_points_y, 'b.', markersize=2, label='LiDAR Scan Points')

# Draw LiDAR rays to detected points or max range
for i in range(num_beams):
    distance = lidar_scan[i, 0]
    angle_offset = lidar_scan[i, 1]

    # Calculate the global angle of the current ray
    ray_global_angle = robot_orientation + angle_offset

    # Calculate the end point of the ray
    ray_end_x = robot_pos[0] + distance * np.cos(ray_global_angle)
    ray_end_y = robot_pos[1] + distance * np.sin(ray_global_angle)

    # Draw the ray
    plt.plot([robot_pos[0], ray_end_x], [robot_pos[1], ray_end_y], 'b-', linewidth=0.2, alpha=0.5)

plt.title('Simulated LiDAR Scan in Virtual Environment')
plt.xlabel('X position (m)')
plt.ylabel('Y position (m)')
plt.legend()
plt.grid(True)
plt.show()
```

## Process LiDAR Data for Obstacle Detection

### Subtask:
Process the simulated LiDAR data to identify and localize obstacles. This could involve techniques like clustering point clouds or filtering to distinguish obstacles from the background.


**Reasoning**:
The first step is to convert the existing polar LiDAR scan data (distance and angle_offset) into Cartesian coordinates (x, y) relative to the robot's position and orientation. This step leverages the robot's current pose to accurately map the detected points in the global environment, creating the 'point cloud' as required by the instructions.



```python
print('--- Step 1: Convert LiDAR scan to Cartesian coordinates ---')

# Combine distance and angle_offset from lidar_scan
distances = lidar_scan[:, 0]
angle_offsets = lidar_scan[:, 1]

# Calculate global angles for each beam
global_angles = robot_orientation + angle_offsets

# Convert polar coordinates to Cartesian coordinates (x, y) relative to the robot
# These are the same calculations used for plotting, but now stored for further processing
point_cloud_x = robot_pos[0] + distances * np.cos(global_angles)
point_cloud_y = robot_pos[1] + distances * np.sin(global_angles)

# Combine x and y coordinates into a single point cloud array
point_cloud = np.vstack((point_cloud_x, point_cloud_y)).T

print(f"Generated point cloud with {len(point_cloud)} points.")
print("First 5 point cloud entries (x, y):")
for i in range(min(5, len(point_cloud))):
    print(f"  X: {point_cloud[i, 0]:.2f}, Y: {point_cloud[i, 1]:.2f}")
```

**Reasoning**:
Now that the LiDAR scan data has been converted to Cartesian coordinates, the next step is to filter out points that represent the maximum range of the sensor rather than actual obstacles. This will refine the point cloud to include only relevant obstacle detections, addressing instruction #2 of the subtask.



```python
print('\n--- Step 2: Filter out max range points ---')

# Filter out points where distance equals max_range
# We'll consider points at max_range as not detecting an obstacle within the scan area
# Using a small epsilon to avoid floating point comparison issues
epsilon = 1e-6 # A small value for floating point comparison

# Create a boolean mask for distances less than max_range
actual_detection_mask = (distances < max_range - epsilon)

# Apply the mask to both distances and the point_cloud
filtered_distances = distances[actual_detection_mask]
filtered_point_cloud = point_cloud[actual_detection_mask]

print(f"Original point cloud size: {len(point_cloud)} points")
print(f"Filtered point cloud size: {len(filtered_point_cloud)} points (excluding max range detections)")

if len(filtered_point_cloud) > 0:
    print("First 5 filtered point cloud entries (x, y):")
    for i in range(min(5, len(filtered_point_cloud))):
        print(f"  X: {filtered_point_cloud[i, 0]:.2f}, Y: {filtered_point_cloud[i, 1]:.2f}")
else:
    print("No obstacles detected within max_range after filtering.")
```

**Reasoning**:
With the filtered point cloud containing only actual obstacle detections, the next step is to implement a clustering algorithm to identify distinct obstacles. This addresses instruction #3 of the subtask by grouping points that are close together into individual obstacle clusters.



```python
print('\n--- Step 3: Cluster filtered points into distinct obstacles ---')

# Define a distance threshold for clustering
cluster_distance_threshold = 1.0 # meters, points closer than this belong to the same cluster

# A simple clustering approach: iterate through points and group nearby ones
# For more complex scenarios, DBSCAN or other clustering algorithms might be used.

obstacle_clusters = []
processed_indices = set()

if len(filtered_point_cloud) > 0:
    # Iterate through each point in the filtered point cloud
    for i in range(len(filtered_point_cloud)):
        if i in processed_indices:
            continue

        # Start a new cluster with the current point
        current_cluster = [filtered_point_cloud[i]]
        processed_indices.add(i)

        # Find all points within the threshold distance to any point in the current cluster
        # This is a breadth-first search like approach to grow clusters
        j = 0
        while j < len(current_cluster):
            current_point = current_cluster[j]
            for k in range(len(filtered_point_cloud)):
                if k in processed_indices:
                    continue

                next_point = filtered_point_cloud[k]
                # Calculate Euclidean distance between current_point and next_point
                distance = np.linalg.norm(current_point - next_point)

                if distance < cluster_distance_threshold:
                    current_cluster.append(next_point)
                    processed_indices.add(k)
            j += 1

        # Convert current_cluster to a numpy array for easier processing later
        if len(current_cluster) > 0:
            obstacle_clusters.append(np.array(current_cluster))

print(f"Identified {len(obstacle_clusters)} distinct obstacle clusters.")
if len(obstacle_clusters) > 0:
    print(f"First cluster has {len(obstacle_clusters[0])} points. Example point: {obstacle_clusters[0][0]}")
```

**Reasoning**:
Now that the LiDAR points have been clustered into distinct obstacles, the final step for this subtask is to calculate the centroid and an estimated size (radius) for each cluster. This will provide a localized position and an approximate dimension for each detected obstacle, fulfilling instruction #4 of the subtask.



```python
print('\n--- Step 4: Calculate centroids and estimated sizes for each obstacle ---')

detected_obstacles_info = [] # List to store dictionaries of obstacle info

if len(obstacle_clusters) > 0:
    for i, cluster in enumerate(obstacle_clusters):
        # Calculate centroid (average x, y coordinates)
        centroid = np.mean(cluster, axis=0)

        # Estimate radius: find the maximum distance from the centroid to any point in the cluster
        # This gives a 'bounding circle' radius for the cluster
        max_dist_from_centroid = 0.0
        if len(cluster) > 0:
            distances_to_centroid = np.linalg.norm(cluster - centroid, axis=1)
            max_dist_from_centroid = np.max(distances_to_centroid)

        # A simple estimation: for visualization and further processing, let's add a small buffer
        # or consider a minimum radius if the cluster is just a single point.
        estimated_radius = max_dist_from_centroid # + some_buffer (optional)
        if estimated_radius < 0.5: # Ensure a minimum visible radius for tiny clusters or single points
            estimated_radius = 0.5

        detected_obstacles_info.append({
            'id': i + 1,
            'centroid': centroid,
            'estimated_radius': estimated_radius,
            'num_points': len(cluster)
        })

    print(f"Identified {len(detected_obstacles_info)} obstacles with centroids and estimated sizes.")
    for obs_info in detected_obstacles_info:
        print(f"  Obstacle {obs_info['id']}: Centroid=({obs_info['centroid'][0]:.2f}, {obs_info['centroid'][1]:.2f}), Radius={obs_info['estimated_radius']:.2f}, Points={obs_info['num_points']}")
else:
    print("No obstacle clusters were found to process.")
```

## Implement Obstacle Prediction Model

### Subtask:
Develop a predictive model to forecast the future positions of detected obstacles. This will involve defining initial velocities for obstacles and creating a function to predict their future positions based on a simple kinematic model.


**Reasoning**:
The first step in implementing the obstacle prediction model is to define the dynamic obstacles, including their initial positions, radii, and assigned velocities (vx, vy), some of which will be zero for static obstacles, as well as the prediction horizon and time step.



```python
print('--- Step 1: Define dynamic obstacles with velocities ---')

# Extend the initial obstacles list with velocity components [vx, vy]
# Obstacles format: [x, y, radius, vx, vy]
dynamic_obstacles = [
    [10, 10, 3, 0.5, 0.2],  # Obstacle 1: Moving
    [25, 20, 5, 0, 0],      # Obstacle 2: Static
    [40, 35, 4, -0.3, 0.4], # Obstacle 3: Moving
    [15, 40, 2, 0, 0]       # Obstacle 4: Static
]

# Define prediction parameters
prediction_horizon_time = 5   # seconds
prediction_time_step = 0.5    # seconds

print("Dynamic Obstacles (x, y, radius, vx, vy):")
for obs in dynamic_obstacles:
    print(f"  - x={obs[0]}, y={obs[1]}, r={obs[2]}, vx={obs[3]}, vy={obs[4]}")
print(f"Prediction Horizon Time: {prediction_horizon_time} seconds")
print(f"Prediction Time Step: {prediction_time_step} seconds")
```

**Reasoning**:
Now that the dynamic obstacles and prediction parameters are defined, the next logical step is to create a function to predict the future positions of a single obstacle based on a simple kinematic model (constant velocity). This directly addresses instruction #3 of the subtask.



```python
print('\n--- Step 2: Create a function to predict obstacle path ---')

def predict_obstacle_path(obstacle, prediction_horizon_time, prediction_time_step):
    x, y, radius, vx, vy = obstacle
    predicted_path = []

    # Calculate number of prediction steps
    num_steps = int(prediction_horizon_time / prediction_time_step)

    # Predict position at each time step
    for i in range(num_steps + 1): # Include the current position (t=0) and all subsequent steps
        current_time = i * prediction_time_step
        predicted_x = x + vx * current_time
        predicted_y = y + vy * current_time
        predicted_path.append(np.array([predicted_x, predicted_y]))

    return np.array(predicted_path)

# Test the prediction function with a single obstacle
test_obstacle = dynamic_obstacles[0] # Take the first dynamic obstacle
test_predicted_path = predict_obstacle_path(test_obstacle, prediction_horizon_time, prediction_time_step)

print(f"Predicted path for obstacle {test_obstacle[0]}, {test_obstacle[1]} (vx={test_obstacle[3]}, vy={test_obstacle[4]}):")
print(f"  Number of points in path: {len(test_predicted_path)}")
print("  First 3 points:")
for i in range(min(3, len(test_predicted_path))):
    print(f"    ({test_predicted_path[i, 0]:.2f}, {test_predicted_path[i, 1]:.2f})")
print("  Last 3 points:")
for i in range(max(0, len(test_predicted_path) - 3), len(test_predicted_path)):
    print(f"    ({test_predicted_path[i, 0]:.2f}, {test_predicted_path[i, 1]:.2f})")
```

**Reasoning**:
Now that the `predict_obstacle_path` function is defined, the next step is to apply it to all `dynamic_obstacles` to generate their predicted paths, as required by instruction #4 of the subtask. This will create a comprehensive list of future trajectories for all identified obstacles.



```python
print('\n--- Step 3: Generate predicted paths for all dynamic obstacles ---')

predicted_obstacle_paths = []

for i, obstacle in enumerate(dynamic_obstacles):
    path = predict_obstacle_path(obstacle, prediction_horizon_time, prediction_time_step)
    predicted_obstacle_paths.append({
        'id': i + 1,
        'initial_state': obstacle, # Store initial x, y, r, vx, vy
        'path': path
    })

print(f"Generated predicted paths for {len(predicted_obstacle_paths)} obstacles.")
if len(predicted_obstacle_paths) > 0:
    print(f"Example: Obstacle {predicted_obstacle_paths[0]['id']} (initial pos: {predicted_obstacle_paths[0]['initial_state'][0]}, {predicted_obstacle_paths[0]['initial_state'][1]}) has {len(predicted_obstacle_paths[0]['path'])} predicted points.")
    print(f"  First point of path: ({predicted_obstacle_paths[0]['path'][0, 0]:.2f}, {predicted_obstacle_paths[0]['path'][0, 1]:.2f})")
    print(f"  Last point of path: ({predicted_obstacle_paths[0]['path'][-1, 0]:.2f}, {predicted_obstacle_paths[0]['path'][-1, 1]:.2f})")
```

## Develop Avoidance Strategy

### Subtask:
Design an algorithm to generate an avoidance path or trajectory for the 'robot' based on the predicted obstacle movements. This will consider the robot's capabilities and the predicted obstacle paths to find a collision-free route.


**Reasoning**:
The first step in developing an avoidance strategy is to define the robot's movement capabilities, including its maximum linear and angular velocities, and a safety radius, as specified in instruction #1 of the subtask.



```python
print('--- Step 1: Define robot movement capabilities ---')

# Robot's movement capabilities
robot_max_linear_velocity = 1.0  # meters per second
robot_max_angular_velocity = np.pi / 4 # radians per second (45 degrees/sec)
robot_safety_radius = 1.0 # meters (buffer zone around the robot)

# Define the robot's goal/target position
robot_target_pos = np.array([45.0, 45.0]) # Example target in the environment

print(f"Robot Max Linear Velocity: {robot_max_linear_velocity} m/s")
print(f"Robot Max Angular Velocity: {np.degrees(robot_max_angular_velocity):.2f} deg/s")
print(f"Robot Safety Radius: {robot_safety_radius} m")
print(f"Robot Target Position: {robot_target_pos}")
```

**Reasoning**:
Now that the robot's capabilities and target are defined, the next logical step is to create a function to check for potential collisions between the robot's current position and any of the predicted obstacle positions, considering their safety radii. This addresses instruction #2 of the subtask.



```python
print('\n--- Step 2: Create a collision checking function ---')

def check_collision(robot_current_pos, robot_safety_radius, obstacle_pos, obstacle_radius):
    # Calculate the distance between the robot's center and the obstacle's center
    distance = np.linalg.norm(robot_current_pos - obstacle_pos)

    # Collision occurs if the distance is less than the sum of their radii
    # (robot_safety_radius + obstacle_radius)
    if distance < (robot_safety_radius + obstacle_radius):
        return True, distance
    else:
        return False, distance

# Test the collision function
# Use robot_pos from 'Simulate Virtual LiDAR Data' section
# Use predicted_obstacle_paths from 'Implement Obstacle Prediction Model' section

# Example: Check collision with the first obstacle at its initial predicted position
if len(predicted_obstacle_paths) > 0:
    test_obstacle_info = predicted_obstacle_paths[0]
    test_obstacle_initial_pos = test_obstacle_info['path'][0] # x, y
    test_obstacle_radius = test_obstacle_info['initial_state'][2] # radius

    is_colliding, dist_to_obs = check_collision(robot_pos, robot_safety_radius, test_obstacle_initial_pos, test_obstacle_radius)

    print(f"Robot at {robot_pos} (safety radius {robot_safety_radius:.2f}m) vs Obstacle 1 at {test_obstacle_initial_pos} (radius {test_obstacle_radius:.2f}m):")
    print(f"  Distance: {dist_to_obs:.2f}m")
    print(f"  Collision expected if distance < {(robot_safety_radius + test_obstacle_radius):.2f}m")
    if is_colliding:
        print("  -> COLLISION DETECTED!")
    else:
        print("  -> No collision.")
else:
    print("No predicted obstacle paths to test collision with.")
```

**Reasoning**:
With the collision checking function in place, the next step is to implement a simple avoidance strategy as described in instruction #3 of the subtask. This involves calculating a new temporary target for the robot when a collision is detected, guiding it away from the obstacle.



```python
print('\n--- Step 3: Implement a simple avoidance strategy ---')

def generate_avoidance_target(robot_current_pos, robot_safety_radius, obstacle_pos, obstacle_radius):
    # Calculate the vector from obstacle to robot
    vector_obs_to_robot = robot_current_pos - obstacle_pos

    # Normalize the vector to get a direction
    if np.linalg.norm(vector_obs_to_robot) == 0: # If robot and obstacle are at the same point
        avoidance_direction = np.array([np.cos(robot_orientation + np.pi/2), np.sin(robot_orientation + np.pi/2)]) # Turn left
    else:
        avoidance_direction = vector_obs_to_robot / np.linalg.norm(vector_obs_to_robot)

    # Calculate a new target position slightly further away from the obstacle
    # This pushes the robot away from the obstacle's current position
    safe_distance = robot_safety_radius + obstacle_radius + 1.0 # Add extra buffer
    new_target = obstacle_pos + avoidance_direction * safe_distance

    # Optionally, steer perpendicular to the collision vector for a more 'around' avoidance
    # For a simple strategy, just moving directly away is a start.

    return new_target

# Test the avoidance strategy with a hypothetical collision
# Let's assume a collision IS detected with the first obstacle at its initial position
# To simulate a collision for testing, let's place the robot very close to the obstacle
hypothetical_robot_pos = np.array([8.0, 8.0]) # Closer to Obstacle 1 (10, 10)

if len(predicted_obstacle_paths) > 0:
    test_obstacle_info = predicted_obstacle_paths[0]
    test_obstacle_current_pos = test_obstacle_info['path'][0] # Use initial predicted pos for simplicity
    test_obstacle_radius = test_obstacle_info['initial_state'][2]

    is_colliding_test, dist_test = check_collision(hypothetical_robot_pos, robot_safety_radius, test_obstacle_current_pos, test_obstacle_radius)

    print(f"Hypothetical Robot at {hypothetical_robot_pos} vs Obstacle 1 at {test_obstacle_current_pos}:")
    print(f"  Distance: {dist_test:.2f}m, Collision expected if distance < {(robot_safety_radius + test_obstacle_radius):.2f}m")

    if is_colliding_test:
        print("  -> COLLISION DETECTED with hypothetical robot. Generating avoidance target...")
        avoid_target = generate_avoidance_target(hypothetical_robot_pos, robot_safety_radius, test_obstacle_current_pos, test_obstacle_radius)
        print(f"  Avoidance Target: ({avoid_target[0]:.2f}, {avoid_target[1]:.2f})")
    else:
        print("  -> No collision with hypothetical robot, no avoidance needed.")
else:
    print("No predicted obstacle paths to test avoidance with.")
```

Create a visualization that displays the virtual LiDAR scans, detected obstacles, predicted obstacle movements, and the planned avoidance path. Ensure all elements are clearly labeled with legends.
